```
## 🛠️ What is a Tool?

A tool is just a normal Python function that the AI is allowed to use.

Example:

add two numbers
count words
search something
call an API
```

## 🛠 Tools used in this notebook

You define simple functions like:
- addition

```
What is an Agent?

An agent is the “system” that connects everything together.

It is:

LLM (brain)
Tools (hands)
Decision logic (when to use tools)
Simple idea:

Agent = AI that can think + choose actions + use tools

🔁 How agent works (real flow)

User asks:

"What is 10 + 5?"

Agent does:

Reads question
Thinks: “This is math”
Chooses tool: add_numbers
Calls tool: add_numbers(10, 5)
Gets result: 15
Returns answer
```

## 🤖 What is a ReAct Agent?

ReAct stands for:
- **Reasoning** (thinking about the problem)
- **Acting** (using tools to solve it)

So the agent:
1. Reads the question
2. Thinks what is needed
3. Uses a tool if required
4. Returns final answer

````python
from dotenv import load_dotenv
from langchain_groq import ChatGroq

load_dotenv()

llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0.7,
)

print("LLM loaded")
````

````python
from langchain_core.tools import tool

@tool
def add_numbers(a: float, b: float) -> float:
    """Add two numbers together."""
    return a + b

@tool
def multiply_numbers(a: float, b: float) -> float:
    """Multiply two numbers."""
    return a * b

@tool
def get_word_count(text: str) -> int:
    """Count number of words in a text."""
    return len(text.split())

@tool
def to_uppercase(text: str) -> str:
    """Convert text to uppercase."""
    return text.upper()

tools = [add_numbers, multiply_numbers, get_word_count, to_uppercase]

print([t.name for t in tools])
````

```
This creates a ReAct agent:

ReAct = Reason + Act
The model:
thinks about the question
decides if it needs a tool
uses the tool
returns final answer

```

````python
from langgraph.prebuilt import create_react_agent

agent = create_react_agent(llm, tools)

print("Agent created 🚀")
````

````python
response = agent.invoke({
    "messages": [("user", "What is 10 + 5 and 3 * 7?")]
})

print(response["messages"][-1].content)
````

````python
response = agent.invoke({
    "messages": [("user", "What is 10 + 5 and 3 * 7?")]
})

print(response["messages"][0].content)
````
